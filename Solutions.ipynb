##### A
- $k_{\text{max}} = n-1$
  - Each node has exactly one edge to every other node
- $k_{\text{min}} = n-1$
  - Same as maximum
- $C= \frac{{n \choose 3}\times 3}{{n \choose 3}\times 3}=1$
  - Every set of 3 nodes forms a triangle, so the number of triangles is the same as the number of connected triples. With $n$ nodes there are ${n \choose 3}$ triples, which we multiply by $3$ to account for the symmetry of each triangle.
- $l_{\text{max}} = 1$
  - The network is not weighted, so each edge has length $1$. Since each vertex has an edge to every other vertex, the longest path between any two vertices is trivially $1$.

##### B
- $k_{\text{max}} = 3$
  - An internal node has exactly $3$ edges (one to its parent and two to its children)
- $k_{\text{min}} = 1$
  - A leaf node has exactly $1$ edge (to its parent)
- $C=\frac{0\times 3}{\frac{3n}{2}-3.5}=0$
  - There are no triangles in a tree. The denominator is the number of connected triples in a tree with $n$ nodes. I arrived at $\frac{3n}{2}$ because at depth > 1, each pair of children adds $3$ triplets. $3.5$ is subtracted because this pattern does not hold for the first two levels of the tree. I arrived at $3.5$ by checking a few levels manually - I'm not sure why it works out to exactly $3.5$.
- $l_{\text{max}} = 2\sqrt{n}$
  - The longest path in a perfect binary tree is any path from a leaf to another leaf on the opposite side of the tree. So we need to travel up to the root, which is $\sqrt{n}$ edges, and then travel back down to the opposite leaf, which is another $\sqrt{n}$ edges.
- $\langle k \rangle = \frac{1}{n} (2+3(\frac{n-1}{2}-1)+(\frac{n+1}{2}))$
  - To find the mean degree, we separate the nodes into three categories. The root has degree $2$, the internal nodes have degree $3$, and the leaf nodes have degree $1$. The tree has $\frac{n-1}{2}-1$ internal nodes and $\frac{n+1}{2}$ leaf nodes, so the mean degree is simply the sum of these nodes divided by $n$.

## Problem 4

In a bipartite network, edges can only connect nodes in $n_1$ to nodes in $n_2$. Therefore, for every edge in the one-mode projection of graph $1$ there is a corresponding edge in the one-mode projection of graph $2$, meaning that the mean degrees are equal. More formally, we calculate the mean degrees as $$c_i = \frac{m_i}{n_i}$$ where $m_i$ is the number of edges in each graph. Solving for $m_i$, we get $$m_i=c_i n_i$$ Because every edge connects a node of type $1$ to a node of type $2$, we know that $$m_1 = m_2$$ Then we can plug in the two sides of the equation to get $$m_1 = m_2 = c_1 n_1 = c_2 n_2$$ Which trivially transforms to $$c_2 = \frac{n_1}{n_2} c_1$$

```python
def read_edgelist(file):
	# make an adjacency list
	adjacency_list = {}
	current_node = 1
	current_adjacencies = []
	with open(file) as f:
		while True:
			# split the line by whitespace
			edge = f.readline().split()
			# if we've reached the end of the file, we break out of the loop
			if not edge:
				adjacency_list[current_node] = current_adjacencies
				break
			# the files are sorted by the second node, so we can build up each node's list before appending it to the full adjacency list
			# this helps avoid excessive read/writes to the large data structure
			if int(edge[1]) == current_node:
				current_adjacencies.append(int(edge[0]))
			else:
				# now we're at the next node, so we append the completed adjacency list for the previous node to the full adjacency list
				adjacency_list[current_node] = current_adjacencies
				# then we reset the current node and adjacency list and start building up the next node's list
				current_node = int(edge[1])
				current_adjacencies = [int(edge[0])]
	
	return adjacency_list

def mean_degree_from_adjacency_list(adj):
	# calculate the mean degree of the graph by couting each node's edges and dividing by the number of nodes
	degrees = [len(adj[node]) for node in adj]
	return sum(degrees) / len(adj)

def plot_mean_degree_distribution_histogram(mean_degrees):
	# plot a simple histogram of the degree distribution
	fig,ax = plt.subplots(figsize=(5,3),dpi=300)
	sns.histplot(mean_degrees,ax=ax,bins=30,edgecolor='black',fill=True,facecolor='black')
	ax.set_title('Mean Degree Distribution')
	ax.set_xlabel(r'Mean Degree $\langle k \rangle$')
	ax.set_ylabel('Frequency')
	sns.despine()
	plt.show()
```

### Part B

We can define the degree of a fixed node $v$ as
$$
k_v = \sum_{u=1}^{n} A_{uv}
$$
which forms a part of our equation for $\langle k_v \rangle$. We can use this to simplify to
$$
\begin{align}
\langle k_v \rangle &= \frac{1}{2m}\sum_{u=1}^{n} \sum_{v=1}^{n} k_v A_{uv} \\
&= \frac{1}{2m}\sum_{v=1}^{n} k_v^2
\end{align}
$$
Recall that the mean degree of a network is defined as
$$
\langle k \rangle = \frac{2m}{n}
$$
Which we can transform to
$$
2m=n\langle k \rangle
$$
Subsituting this into our equation for $\langle k_v \rangle$ gives us
$$
\langle k_v \rangle = \frac{1}{n\langle k \rangle}\sum_{v=1}^{n} k_v^2
$$
Note that
$$
\frac{1}{n}\sum_{v=1}^{n} k_v^2 = \langle k^2 \rangle
$$
So we can simplify to
$$
\langle k_v \rangle = \frac{\langle k^2 \rangle}{\langle k \rangle}
$$

Very cool!

```python
# Scatterplot of the size of the paradox (mean of k_v / mean degree of k_u) vs mean degree k_u

def mean_squared_degree_from_adjacency_list(adj):
	# calculate the mean degree of the graph by couting each node's edges and dividing by the number of nodes
	squared_degrees = [len(adj[node])**2 for node in adj]
	return sum(squared_degrees) / len(adj)

def get_xy(adj):
	# find the mean degree and mean squared degree using simple helper functions
	mean_k = mean_degree_from_adjacency_list(adj)
	mean_k_squared = mean_squared_degree_from_adjacency_list(adj)
	# find the mean neighbor degree using the equation from part B
	mean_kv = mean_k_squared / mean_k
	# calculate the paradox size
	paradox_size = mean_kv / mean_k

	return mean_k, paradox_size

```

```python
mnd_df = pd.DataFrame(columns=['edgelist_file','school','mean_k','paradox_size'])

for file in edgelist_files:
	# get the school name with regex
	school = re.match(r'([a-zA-Z]+)',os.path.basename(file)).group(0)
	# get the adjacency list
	adj = read_edgelist(file)
	# get the mean degree and paradox size
	mean_k, paradox_size = get_xy(adj)
	# add the data to the dataframe
	mnd_df.loc[len(mnd_df)] = [file,school,mean_k,paradox_size]
	print('.',end='')
```

### Part D
The majority illusion can occur when nodes of disproportionately high degree frequently exhibit the property of interest. The friendship paradox tells us that most nodes will tend to have neighbors with more friends, so if the property is exhibited by high-degree nodes, we can end up with the majority of nodes (which tend to have smaller degree) experiencing the majority illusion. In other words, most nodes will have few friends, so requires relatively few friends to form a majority. In a network with high-degree nodes exhibiting the property, it's possible for the majority of nodes to experience this illusion despite most nodes not having the property. This becomes less likely as the variance in degrees decreases and as $q$ decreases further below 0.5.

For example, we could construct a network with odd $n$ in which all nodes with $k> \langle k\rangle$ have the property (so $q<0.5$). The friendship paradox tells us that $\langle k_v \rangle > \langle k\rangle$, and in this network, majority of the average node's friends will have $k> \langle k\rangle$ and will thereby also exhibit the property. This creates a perceived prevalence of the property greater than 0.5 despite $q<0.5$.

### Part E

```python
from collections import deque

def breadth_first_search(adj, start):
    queue = deque([start])
    visited = {start: 0}
    while queue:
        node = queue.popleft()
        for neighbor in adj[node]:
            if neighbor not in visited:
                visited[neighbor] = visited[node] + 1
                queue.append(neighbor)
    return visited

def find_largest_component(adj):
    visited = set()
    largest_component = set()
    for node in adj:
        if node not in visited:
            component = breadth_first_search(adj, node).keys()
            if len(component) > len(largest_component):
                largest_component = component
            visited.update(component)
    return largest_component

def diameter_from_adjacency_list(adj):
    largest_component = find_largest_component(adj)
    if not largest_component:
        return 0, 0

    # Run BFS from an arbitrary node in the largest component
    start_node = next(iter(largest_component))
    bfs_result = breadth_first_search(adj, start_node)
    
    # Find the farthest node from the start_node
    farthest_node = max(bfs_result, key=bfs_result.get)
    
    # Run BFS again from the farthest node to find the diameter
    bfs_result = breadth_first_search(adj, farthest_node)
    diameter = max(bfs_result.values())

    # Calculate the mean geodesic distance
    total_distance = sum(bfs_result.values())
    mean_geodesic_distance = total_distance / len(largest_component)

    return int(diameter), mean_geodesic_distance, len(largest_component)
```

```python
# run on all networks
for i,r in mnd_df.iterrows():
	adj = read_edgelist(r['edgelist_file'])
	mnd_df.loc[i,'diameter'], mnd_df.loc[i,'mean_geodesic_distance'], mnd_df.loc[i,'largest_component_size'] = diameter_from_adjacency_list(adj)
	mnd_df.loc[i,'size'] = len(adj)
	print('.',end='')

mnd_df['n_edges'] = mnd_df['mean_k'] * mnd_df['size'] / 2
```
